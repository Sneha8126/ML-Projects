```python
df['corpus'] = df['corpus'].fillna('')
```

```python
selected_features = ['corpus']
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
feature_vectors = vectorizer.fit_transform(df['corpus'])
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
mobile_name = input("Enter Mobile Name : ")
```

```python
all_mobile_names = df['name'].tolist()
```

```python
find_close_match = difflib.get_close_matches(
    mobile_name,
    all_mobile_names
)
```

```python
close_match = find_close_match[0]
```

```python
print(df.columns)
```

```python
index = df[df['name'] == close_match].index[0]
```

```python
df = df.reset_index()
```

```python
similarity_score = list(enumerate(similarity[index]))
```

```python
sorted_similar_mobiles = sorted(
    similarity_score,
    key=lambda x: x[1],
    reverse=True
)
```

```python
print("Recommended Mobiles:\n")

i = 1

for mobile in sorted_similar_mobiles:

    index = mobile[0]

    title = df[df.index == index]['name'].values[0]

    if i < 11:
        print(i, ".", title)
        i += 1
```

```python
print("Recommended Mobiles\n")

i = 1

for mobile in sorted_similar_mobiles:

    index = mobile[0]

    row = df.iloc[index]

    if i <= 10:

        print("-----------------------------")
        print("Mobile :", row['name'])
        print("Rating :", row['ratings'])
        print("Price :", row['price'])
        print("-----------------------------")

        i += 1
```
